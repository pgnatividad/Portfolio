```python
song_list = []
for song in albums_n_songs.find_all('a'):
    track = {}
    song_url = song.attrs['href']
    song_url = song_url[2:]
    song_res =requests.get("https://www.azlyrics.com"+song_url)
    soup2 = BeautifulSoup(song_res.content,'lxml')
    lyric_tags=soup2.find('div',{'class':'col-xs-12 col-lg-8 text-center'})
    test2 = lyric_tags.find_all('div')
    track['lyrics']= test2[6]
    track['title'] = song.text
    time.sleep(10)
    song_list.append(track)
    
song_list
```
